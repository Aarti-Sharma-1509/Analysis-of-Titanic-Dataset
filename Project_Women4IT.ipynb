# Cleaning of merged dataset

```python
## determining the duplicate rows
## Rows containing duplicate data

duplicate_rows = combined[combined.duplicated()]
print("number of duplicate rows are : ", duplicate_rows.shape)
```

```python
# Dropping the duplicates 

combined = combined.drop_duplicates()
combined.head(5)
```

```python
# Finding the null values.
print(combined.isnull().sum())
```

```python
##filling the missing values

## filling the values in age column by their mean value
combinedA['Age'] = combinedA['Age'].fillna(combinedA['Age'].mean())

## filling the values in Embarked column by 'S'
combinedA['Embarked'] = combinedA['Embarked'].fillna('C')

## filling the values in Fare column by their mean value
combinedA['Fare'] = combinedA['Fare'].fillna(combinedA['Fare'].mean())

## filling the values in Survived column by '0'

combinedA['Survived'] = combinedA['Survived'].fillna(0)
```

# Machine learning algorithms for Decision Tree

```python
##  Dropping non required columns

combinedN=combinedA.drop(['PassengerId','Name','Ticket'], axis=1)
combinedN.shape
```

```python
## converting string into categorical type for sex column

combinedN["Sex"] = combinedN["Sex"].astype('category')   ## where male=1 and female =0
combinedN["Sex_cat"] = combinedN["Sex"].cat.codes

combinedN["Embarked"] = combinedN["Embarked"].astype('category')     ##### where Q=1, S=2, C=3
combinedN["Embarked_cat"] = combinedN["Embarked"].cat.codes

print(combinedN.dtypes)
combinedN.head()
```

```python
### Dropping Sex and Embarked column as same values has benn stored in other columns in form of categorical values
# Dropping the missing values.

combinedN = combinedN.drop(["Sex"],axis=1)  ### dropped column "Sex"
combinedN= combinedN.drop(["Embarked"],axis=1)
combinedN
```

```python
##  splitting into test data and train data to target feature

x = combinedN.drop("Survived", axis=1)   ### train data
y = combinedN["Survived"]               ### test data
```

```python
## Split the data into training and validation sets as 75:25

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.25,random_state=1)
```

```python
## Take a look at the shape of splitted data

x_train.shape, y_train.shape
```

```python
## fitting decision tree for trained dataset

decision_tree = DecisionTreeClassifier(max_depth = 2,random_state=1) 
decision_tree.fit(x_train, y_train)  
```

```python
## predicted values of the test data

y_pred = decision_tree.predict(x_test)  
y_pred       ### prediction score for target
```

```python
## plotting decision tree

tree.plot_tree(decision_tree,max_depth = 3,
               impurity = True,
               feature_names = list(combinedN.drop(['Survived'], axis=1)),
               class_names = ['Died', 'Survived'],
               rounded = True, 
               filled= True );
```

```python
## run upto 5 fold cross validation
## cross validation 

CV = cross_val_score(decision_tree,x,y,cv=5)
print("cross validation score upto 5th fold :", CV)
```

```python
### Accuracy with CV score of decision tree

print("Accuracy: " + str((CV.mean()*100.0))+ "And  with std: "+ str(CV.std()*100.0))
```

## Performance evaluation

```python
## accuracy for the decision tree for test data of combined dataset

accuacy = round(decision_tree.score(x_test, y_test) * 100, 2)
accuacy

##  Thus, It can be conclude that 81.10 % of cross-validation set tuples were correctly classified by this model.
```

```python
## Take a look at the confusion matrix ([TN,FN],[FP,TP])

confusion_matrix=confusion_matrix(y_test,y_pred)
confusion_matrix
```

```python
## Precision score

Precision=("Precision score: %0.4f" % precision_score(y_test, y_pred))
Precision

### Thus, It can be conclude that 70.00 % of tuples that the classifier labeled as positive 
## those are actually positive by this model.
```

```python
## Recall score

Recall=("Recall score: %0.4f" % recall_score(y_test, y_pred))
Recall

##Thus, It can be conclude that 48.84 % of real positive tuples were classified by the decision tree classifier.
```
